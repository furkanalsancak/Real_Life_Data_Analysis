```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('company_sales_data.csv')
```

```python
monthList = data['month_number']
profitList = data['total_profit']
face_cream = data['facecream']
face_wash = data['facewash']
tooth_paste = data['toothpaste']
bathing_soap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']
```

```python
plt.plot(monthList,profitList, label = 'Month-wise Profit data of last year')

plt.xticks(monthList)
plt.yticks([100000,200000,300000,400000,500000])

plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')

plt.title('Company Profit per Month')

plt.tight_layout()
plt.show()
```

```python
plt.plot(monthList,profitList, 
        label = 'Profit data of last year', 
        linestyle='--', 
        linewidth=3, 
        color='red',
        marker='o',
        markerfacecolor='k')

plt.xticks(monthList)
plt.yticks([100000,200000,300000,400000,500000])

plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')

plt.title('Company Profit per Month')

plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

```python
#plt.plot(monthList,profitList, label = 'Month-wise Profit data of last year')
plt.plot(monthList,face_cream, label = 'Face Cream Sales Data'    , linewidth=3 ,marker='o')
plt.plot(monthList,face_wash, label = 'Face Wash Sales Data'      , linewidth=3 ,marker='o')
plt.plot(monthList,tooth_paste, label = 'Toothpaste Sales Data'   , linewidth=3 ,marker='o')
plt.plot(monthList,bathing_soap, label = 'Bathing Soap Sales Data', linewidth=3 ,marker='o')
plt.plot(monthList,shampoo, label = 'Shampoo Sales Data'          , linewidth=3 ,marker='o')
plt.plot(monthList,moisturizer, label = 'Moisturizer Sales Data'  , linewidth=3 ,marker='o')



plt.xticks(monthList)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')

plt.title('Sales Data')

plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.scatter(monthList, tooth_paste, label='Toothpaste Sales Data')

plt.xticks(monthList)
plt.yticks([4500,5000,5500,6000,6500,7000,7500,8000])

plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Tooth paste Sales Data')
plt.grid(True, linewidth= 1, linestyle="--")

plt.legend(loc='upper left')
plt.show()
```

```python

x_indexes = np.arange(len(monthList))
width = 0.25

plt.bar(monthList - 0.13, face_cream, width=width, color='#4b64d6', label='Face Cream Sales Data')
plt.bar(monthList + 0.13, face_wash, width=width, color='#f07b22' , label='Face Wash Sales Data')

plt.xticks(monthList)

plt.grid(True, linewidth= 1, linestyle="--")
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.title('Facewash and Facecream sales data')

plt.legend(loc='upper left', fontsize='11.7')


plt.show()
```

```python
plt.bar(monthList, bathing_soap)

plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.title('Bathing Soap sales data')
```

```python

profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList,profit_range, label = 'Profit data')

plt.title('Profit data')
plt.legend(loc='upper left')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.xticks(profit_range, size=10)
plt.show()
```

```python
labels = ['FaceWash', 'ToothPaste', 'FaceCream', 'Bathing soap', 'Shampoo', 'Moisturizer']

plt.pie([face_wash.sum(), tooth_paste.sum(),
        face_cream.sum(), bathing_soap.sum(), shampoo.sum(), moisturizer.sum()],
        labels=labels, autopct='%1.1f%%')

plt.legend(loc=(1,1), fontsize=10)
plt.title('Sales data')

plt.show()

```

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(monthList, bathing_soap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(monthList, face_wash, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

```python
plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, face_cream, face_wash, tooth_paste, 
              bathing_soap, shampoo, moisturizer, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left', fontsize=10)
plt.show()
```
